# データの可視化

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

main_color = '#fe7e39'
sub_color = '#18216e'
```

```python
df = pd.read_csv('../datasets/preprocessed_data.csv')
display(df.head())
display(df.describe(include='all'))
```

## Age Histgram

```python
fig, axes = plt.subplots(figsize=(11, 5))

ax0 = axes

tmp = df.age 

ax0.text(x=tmp.min(), y=1500, s=f"min: {round(tmp.min(), 1)}")

ax0.hist(tmp, bins=10, color=main_color, label='Age')
ax0.vlines(x=np.mean(tmp), ymin=0, ymax=9000, linewidth=3, linestyles='dotted', color=sub_color, label='Average')
ax0.text(x=np.mean(tmp)*1.01, y=9100, s=round(np.mean(tmp), 1), color=sub_color)

ax0.vlines(x=np.nanquantile(tmp, q=0.5), ymin=0, ymax=9000, linewidth=3, linestyles='dashed', color=sub_color, label='Median')
ax0.text(x=np.nanquantile(tmp, q=0.5)*0.95, y=9100, s=round(np.nanquantile(tmp, q=0.5), 1), color=sub_color)

ax0.text(x=tmp.max()-5, y=1500, s=f"Max: {round(tmp.max(), 1)}")

ax0.minorticks_on()
ax0.set_title('Age Histgram')
ax0.spines[['right', 'top']].set_visible(False)

ax0.legend()
```

## Saving Histgram

```python
fig, axes = plt.subplots(figsize=(11, 10), nrows=2)

ax0, ax1 = axes

tmp = df.saving 
ax0.text(x=tmp.min(), y=1500, s=f"min: {round(tmp.min(), )}")

ax0.hist(tmp, bins=50, color=main_color, label='Saving')
ax0.vlines(x=np.mean(tmp), ymin=0, ymax=25000, linewidth=2, linestyles='dotted', color=sub_color, label='Average')
ax0.text(x=np.mean(tmp)*1.3, y=15000, s=round(np.mean(tmp), 1), color=sub_color)

ax0.vlines(x=np.nanquantile(tmp, q=0.5), ymin=0, ymax=25000, linewidth=2, linestyles='dashed', color=sub_color, label='Median')
ax0.text(x=np.nanquantile(tmp, q=0.5)*5, y=9000, s=round(np.nanquantile(tmp, q=0.5), 1), color=sub_color)

ax0.text(x=tmp.max()-5, y=1500, s=f"Max: {round(tmp.max(), )}")

ax0.set_xlim(0, )
ax0.minorticks_on()
ax0.set_title('Saving Histgram')
ax0.spines[['right', 'top']].set_visible(False)
ax0.legend()

ax1.text(x=tmp.min(), y=1500, s=f"min: {round(tmp.min(), )}")

ax1.hist(tmp, bins=250, color=main_color, label='Saving')
ax1.vlines(x=np.mean(tmp), ymin=0, ymax=17500, linewidth=2, linestyles='dotted', color=sub_color, label='Average')
ax1.text(x=np.mean(tmp)+100, y=15000, s=round(np.mean(tmp), 1), color=sub_color)

ax1.vlines(x=np.nanquantile(tmp, q=0.5), ymin=0, ymax=17500, linewidth=2, linestyles='dashed', color=sub_color, label='Median')
ax1.text(x=np.nanquantile(tmp, q=0.5)+100, y=9000, s=round(np.nanquantile(tmp, q=0.5), 1), color=sub_color)

# ax1.text(x=tmp.max()-5, y=1500, s=f"Max: {round(tmp.max(), )}")

ax1.set_xlim(0, 4000)
ax1.minorticks_on()
ax1.set_title('Saving Histgram')
ax1.spines[['right', 'top']].set_visible(False)
ax1.legend()

plt.tight_layout()
```

## Debt Histgram

```python
fig, axes = plt.subplots(figsize=(11, 10), nrows=2)

ax0, ax1 = axes

tmp = df.debt 
ax0.text(x=tmp.min()+50, y=400, s=f"min: {round(tmp.min(), )}")

ax0.hist(tmp, bins=50, color=main_color, label='Saving')
ax0.vlines(x=np.mean(tmp), ymin=0, ymax=600, linewidth=2, linestyles='dotted', color=sub_color, label='Average')
ax0.text(x=np.mean(tmp)-300, y=600, s=round(np.mean(tmp), 1), color=sub_color)

ax0.vlines(x=np.nanquantile(tmp, q=0.5), ymin=0, ymax=600, linewidth=2, linestyles='dashed', color=sub_color, label='Median')
ax0.text(x=np.nanquantile(tmp, q=0.5)+100, y=400, s=round(np.nanquantile(tmp, q=0.5), 1), color=sub_color)

ax0.text(x=tmp.max()-5, y=200, s=f"Max: {round(tmp.max(), )}")

ax0.set_xlim(tmp.min(), 0)
ax0.minorticks_on()
ax0.set_title('Debt Histgram')
ax0.spines[['right', 'top']].set_visible(False)
ax0.legend()

ax1.hist(tmp, bins=50, color=main_color, label='Saving')
ax1.vlines(x=np.mean(tmp), ymin=0, ymax=600, linewidth=2, linestyles='dotted', color=sub_color, label='Average')
ax1.text(x=np.mean(tmp)-30, y=600, s=round(np.mean(tmp), 1), color=sub_color)

ax1.vlines(x=np.nanquantile(tmp, q=0.5), ymin=0, ymax=600, linewidth=2, linestyles='dashed', color=sub_color, label='Median')
ax1.text(x=np.nanquantile(tmp, q=0.5)+53, y=550, s=round(np.nanquantile(tmp, q=0.5), 1), color=sub_color)

ax1.text(x=tmp.max()-5, y=200, s=f"Max: {round(tmp.max(), )}")

ax1.set_xlim(-2000, 0)
ax1.minorticks_on()
ax1.set_title('Debt Histgram')
ax1.spines[['right', 'top']].set_visible(False)
ax1.legend()

plt.tight_layout()
```
